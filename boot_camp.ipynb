```python
input_file = 'students.txt'

students_data = [
    # 학생이름, 학년, 국, 영, 수 점수
    ['alice', 10, 95, 88, 92],
    ['bob', 8, 90, 58, 42],
    ['david', 6, 99, 88, 99]
]
```

```python
# 1. 학생데이터를 파일로 저장해보자

def write_students(path, datas):

    with open(path, 'w', encoding='utf-8') as file:
        for student in datas:
            tmp = ','.join(map(str, student))
            file.write(tmp+'\n')

write_students(input_file, students_data)
```

```python
# 2. 학생 데이터 읽기
# [{'name':'alice', 'grade':10, 'scores':{'kor':95, 'eng':88, 'math':92}},
#  {'name':'bob',.......}

def read_students(path):

    tmp_students = []
    with open(path, 'r', encoding='utf-8') as file:
        for line in file:
            name, grade, kor, eng, math = line.strip().split(',')
            student = {
                'name': name,
                'grade': grade,
                'scores':{
                    'kor':int(kor),
                    'eng':int(eng),
                    'math':int(math)
                }
            }
            tmp_students.append(student)
    return tmp_students

tmp_s = read_students(input_file)
tmp_s
```

```python
# 3. 읽어온 학생데이터 처리
# 각 학생들의 총점과 평균을 기존 딕셔너리에 추가

def proces_students(datas):

    tmp = []
    for student in datas:
        total_score = sum(student['scores'].values())
        avg_score = total_score / len(student['scores'])
        student['total_score'] = total_score
        student['avg_score'] = round(avg_score,2)
        tmp.append(student)
    return tmp

proces_students(tmp_s)
```

```python
#Q1.
products_list = []
price_list = []

# q가 입력될때까지 무한반복
while True:
    product = input("상품명을 입력하세요 (종료하려면 'q' 입력): ")
    if product == 'q':
        break

    try:
        price = int(input(f"{product}의 가격을 입력하세요: "))
    except ValueError:
        print("가격은 숫자로 입력해주세요. 다시 입력합니다.")
        continue

    products_list.append(product)
    price_list.append(price)

# 입력받은 상품과 가격을 출력
print("\n--- 입력된 상품 목록 ---")
for i in range(len(products_list)):
    print(f"상품: {products_list[i]}, 가격: {price_list[i]}")

# 가격 총합계 출력
total_price = sum(price_list)
print(f"총 합계: {total_price}원")
```

```python
# Q2.
# 반복을 통한 각 과목별 점수 입력
# 점수 평균
input_name=input("이름입력: ")
subject_list=['kor','math','history']
subject_score_list=[]
for subject in subject_list:
  while True:
    try:
      subject_score=int(input(f"{subject} 성적을 입력하세요:"))
      if subject_score > 100 or subject_score < 0:
        print("점수는 0점에서 100점 사이로 입력해주세요.")
        continue
      break
    except ValueError:
      print("성적은 숫자로 입력해주세요. 다시 입력하세요. ")
      continue
  subject_score_list.append(subject_score)

print(f'{input_name}의 성적')
for subject,score in zip(subject_list,subject_score_list):
  print(f'{subject}:{score}')
print('------------')
print(f'평균: {sum(subject_score_list)/len(subject_score_list):.2f}')

```

```python
#Q3.
import random
target = random.randint(1, 101)
attempts_count=0
print('1에서 100까지의 숫자를 맞추어 보세요')
while True:
  try:
    answer=int(input('숫자를 입력하세요:'))
    attempts_count+=1
  except ValueError:
    print('숫자만 입력가능합니다.')
    attempts_count-=1
    continue
  if answer>100 or answer<1:
    print('1에서 100사이의 숫자를 입력하세요')
    attempts_count-=1
    continue
  elif answer > target:
    print("'DOWN'")
  elif answer < target:
    print("'UP'")
  else:
    print(f"'정답입니다.',{attempts_count}번 만에 맞추셨습니다.")
    break
```

```python
from collections import Counter

# Ensure the built-in sum function is used, in case it was overwritten.
# This is a common issue in interactive environments if a variable 'sum' was defined.
if 'sum' in globals() and isinstance(sum, int):
    del sum

vowels='aeiou'
matrix = [
    ['apple', 'banana', 'cherry'],
    ['date', 'fig', 'grape'],
    ['kiwi', 'lemon', 'mango']
]
for row in matrix:
  for word in row:
    res_cnt=Counter([w for w in word if w not in vowels])
    total_cnt=sum(res_cnt.values())
    print(f'단어 {word}의 자음 개수는 {total_cnt}')
```
